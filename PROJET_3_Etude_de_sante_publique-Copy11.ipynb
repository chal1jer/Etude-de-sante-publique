### Calcul du pourcentage de protéines de chaque produit (pour chaque pays). Ce pourcentage est obtenu en calculant le ratio "poids de protéines/poids total" (attention aux unités utilisées).

Teneur en protéines de l'avoine : 10.70g par 100g soit 10.7%  
Aliment le plus riche en protéines : la gélatine alimentaire = 87.60 %. Suivi par la levure alimentaire : 48 %. Vu sur https://sante.journaldesfemmes.fr/calories/classement/aliments/proteines

**Calcul :**  
% de protéines de chaque produit (pour chaque pays) = (dispo ali (kg de protéines/pays/an) / (nourriture x 1 000 000)) x 100  
*La nourriture est exprimée en milliers de tonnes. Dans le calcul, la multiplication par 1 000 000 permet d'avoir la nourriture en kg.  
La multiplication par 100 permet d'avoir le pourcentage.* 

```python
# Création de la colonne "% prot".  
datapop17['% prot'] =  datapop17['dispo ali (kg de protéines/pays/an)'] / (datapop17['nourriture'] * 1000000) * 100
```
