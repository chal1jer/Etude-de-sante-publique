```python
pop.head()
```

# <font color='#FFCA36'>Table "sous_nutrition"</font>

Une table appelée sous_nutrition, contenant le nombre de personnes en sous-alimentation pour chaque pays et pour chaque année comprise entre 2012 et la dernière année disponible. Elle devra contenir 4 colonnes : pays, code_pays, année, nb_personnes.

**Une proposition de clé primaire :** code-pays - année

```python
sous_nut = pd.read_csv("C:/Users/Zang/Documents/APPRENDER/OPENCLASSROOMS/PARCOURS DA/P3_Etude_de_sante_publique/bdd/sous_nutrition12-18.csv")
```

```python
sous_nut.head(2)
```

```python
# Je conserve les colonnes utiles 
sous_nut = sous_nut[['Code zone','Zone','Année','Valeur']]
```

```python
sous_nut.head()
```

##### Colonne Valeur

```python
# Vérification de la colonne pq la multiplication par 1000000 ne fonctionne pas : 
sous_nut['Valeur'].describe()
```

```python
# Vérification de la donnée "<0.1"
sous_nut[sous_nut['Valeur']=='<0.1'].head()
```

```python
# Je remplace toutes les valeurs '<0.1' par 0
sous_nut = sous_nut.replace('<0.1', '0')
```

```python
# Changement du type de la colonne "Valeur" pour pouvoir la multiplier
sous_nut['Valeur'] = sous_nut['Valeur'].astype('float64')
```

```python
sous_nut['Valeur'] *= 1000000
```

```python
# Vérification
sous_nut.head()
```

La transfo de "Valeur" en int est impossible : ValueError: Cannot convert non-finite values (NA or inf) to integer.

```python
# Je cherche quel pays peut être "NA"
sous_nut[sous_nut['pays']=='Allemagne']
```

```python
# Je remplace NaN par 0 : 
sous_nut['nb_personnes'] = sous_nut['nb_personnes'].fillna(0)
```

```python
# Vérification
sous_nut[sous_nut['pays']=='Allemagne']
```

```python
# Je peux changer le type 
sous_nut['nb_personnes'] = sous_nut['nb_personnes'].astype('int64')
```

```python
# Vérification
sous_nut.dtypes
```

```python
# Je remplace les périodes par des année
sous_nut = sous_nut.replace(['2011-2013','2012-2014','2013-2015','2014-2016','2015-2017','2016-2018'], 
                                ['2012','2013','2014','2015','2016','2017'])
```

```python
sous_nut.head()
```

```python
sous_nut.shape
```

```python
# J'enlève les doublons de la Chine 
sous_nut = sous_nut[(sous_nut.Zone != 'Chine - RAS de Hong-Kong') & (sous_nut.Zone != 'Chine - RAS de Macao') & (sous_nut.Zone != 'Chine, continentale') & (sous_nut.Zone != 'Chine, Taiwan Province de')]
```

```python
# Je renomme les colonnes
sous_nut = sous_nut.rename(columns={'Code zone': 'code_pays', 'Zone': 'pays', 'Année': 'année', 'Valeur': 'nb_personnes'})
```
